```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate([
    ("system", "You're a language expert. Translate the given text {text} from {source_language} to {target_language}"),
    ("human","{text}"),
    ("ai", "Thanks for using Ritik Chatgpt")
])

print(prompt_template)
messages = prompt_template.invoke({
    "text": "Hi, My name is Ritik",
    "source_language": "English",
    "target_language": "german"
})
print(messages)

result = model.invoke(messages)
result

```

#### Building Chain

```python
from langchain_core.runnables import RunnableLambda
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser


def analyze_story_chain():
    """
    # equivalent expanded execution (conceptual)
messages = story_prompt.invoke({"theme": "AI", "character": "AI robot"})
model_resp = model.invoke(messages)
story_text = StrOutputParser().invoke(model_resp)

# optional transform
inputs_for_analysis = analyzeStory(story_text)  # -> {"story": story_text}

analysis_messages = analyze_story_prompt.invoke(inputs_for_analysis)
analysis_resp = model.invoke(analysis_messages)
final_analysis = StrOutputParser().invoke(analysis_resp)
    """

    
    story_prompt = ChatPromptTemplate([
        ("system", "You're a great story teller. create a story based on the given theme: {theme} and given character: {character}"),
        ("human", "Theme:{theme}\n Character:{character}")
    ])
    analyze_story_prompt = ChatPromptTemplate([
        ("system", "You're an great lecturer. Review the given story and analyze the story"),
        ("human", "{story}")
    ])

    story_chain = story_prompt | model | StrOutputParser()
    def analyzeStory(story_text):
        return {"story": story_text}

    return (
        story_chain
        | RunnableLambda(analyzeStory)
        | analyze_story_prompt
        | model
        | StrOutputParser()
    )


```

```python
analysisChain = analyze_story_chain()
analysisMessages = analysisChain.invoke({
    "theme": "Artifical Intelligence",
    "character": "AI robot"
})
analysisMessages
```
